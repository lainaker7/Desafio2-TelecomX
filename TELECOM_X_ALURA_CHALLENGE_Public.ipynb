# üìå **1. E-xtracci√≥n de los Datos**: üìå

```python
# Primer paso es Generar la Extracci√≥n de los datos

#üìå Enlace de la API:
#üîó challenge2-data-science-LATAM/TelecomX_Data.json at main ¬∑ ingridcristh/challenge2-data-science-LATAM

import pandas as pd
```

```python
# Si el JSON es directamente un array de objetos
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
df = pd.read_json(url)

# Mostrar las primeras filas
df.head()
```

```python

import requests

# Obtener los datos directamente como diccionario
response = requests.get(url)
data = response.json()

# Normalizar el JSON completo
df_normalized = pd.json_normalize(data)

print("‚úÖ Datos normalizados exitosamente")
print(f"üìä Forma: {df_normalized.shape}")
print("\nüîç Columnas disponibles:")
for col in df_normalized.columns:
    print(f"  - {col}")

print("\nüìã Primeras 5 filas:")
display(df_normalized.head())

print("\nüìä Informaci√≥n del DataFrame:")
print(df_normalized.info())
```
