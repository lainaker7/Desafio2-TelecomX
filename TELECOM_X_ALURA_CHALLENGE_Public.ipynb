```python
# An√°lisis r√°pido de calidad de datos
def quick_data_quality_check(df):
    print("üîç AN√ÅLISIS R√ÅPIDO DE CALIDAD DE DATOS")
    print("="*60)

    results = {}

    # 1. Valores nulos
    null_sum = df.isnull().sum().sum()
    results['null_values'] = null_sum
    print(f"‚úÖ Valores nulos: {null_sum}")

    # 2. Duplicados
    duplicates = df.duplicated().sum()
    results['duplicates'] = duplicates
    print(f"‚úÖ Duplicados completos: {duplicates}")

    # 3. Duplicados en ID
    if 'customerID' in df.columns:
        id_duplicates = df['customerID'].duplicated().sum()
        results['id_duplicates'] = id_duplicates
        print(f"‚úÖ Duplicados en customerID: {id_duplicates}")

    # 4. Tipos de datos incorrectos
    if 'account.Charges.Total' in df.columns:
        if df['account.Charges.Total'].dtype == 'object':
            results['format_errors'] = "Total Charges es object, deber√≠a ser num√©rico"
            print(f"‚ö†Ô∏è  account.Charges.Total: tipo incorrecto (object)")

    # 5. Rangos imposibles
    if 'customer.tenure' in df.columns:
        negative_tenure = (df['customer.tenure'] < 0).sum()
        results['negative_tenure'] = negative_tenure
        if negative_tenure > 0:
            print(f"‚ö†Ô∏è  tenure negativo: {negative_tenure} registros")

    # 6. Valores categ√≥ricos inesperados
    unexpected_counts = {}
    expected_churn = ['Yes', 'No']
    if 'Churn' in df.columns:
        unexpected_churn = set(df['Churn'].unique()) - set(expected_churn)
        if unexpected_churn:
            results['unexpected_churn'] = unexpected_churn
            print(f"‚ö†Ô∏è  Valores inesperados en Churn: {unexpected_churn}")

    print("="*60)
    return results

# Ejecutar an√°lisis r√°pido
quality_report = quick_data_quality_check(df)
```

Viene la parte en la que hay que eliminar nombres repetitivos, ver si hay datos vacios dentro del DataFrame y corregir esas inconsistencias

```python
# Prefijos a eliminar
prefijos = ['customer.', 'phone.', 'internet.', 'account.', 'Charges.']

# Listas para el mapeo (emparejamiento)
nombres_actuales = []
nombres_limpios = []

# Iterar sobre las columnas del DataFrame
for col in df.columns:
    es_prefijo = False

    # Verificar si la columna comienza con alguno de los prefijos
    for prefijo in prefijos:
        if col.startswith(prefijo):
            nombres_actuales.append(col)
            # Eliminar el prefijo para obtener el nombre limpio
            nombres_limpios.append(col.replace(prefijo, '', 1))
            es_prefijo = True
            break

    # Si la columna no tiene prefijo (ej: customerID, Churn), se mantiene igual
    if not es_prefijo:
        nombres_actuales.append(col)
        nombres_limpios.append(col)


# Usar zip() para crear el diccionario de mapeo
# Empareja (zip) los nombres actuales con los nombres limpios
mapeo_renombre = dict(zip(nombres_actuales, nombres_limpios))

# Renombrado con .rename()
# Le pasamos el diccionario de mapeo generado por zip
df = df.rename(columns=mapeo_renombre)
df.info()
```
